# KMeans Clustering

## Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Dataset

```python
df = pd.read_csv("datasets/Mall_Customers.csv")
df
```

```python
X = df.iloc[:, [3,4]].values  # taking only last two features for visualizing purpose
```

## Elbow Method

```python
from sklearn.cluster import KMeans
WCSS = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, init='k-means++',random_state=24)
    kmeans.fit(X)
    WCSS.append(kmeans.inertia_)
sns.set_style(style="darkgrid")
plt.plot(range(1,10),WCSS)
plt.xlabel("# of Clusters")
plt.ylabel("WCSS")
plt.show()
```

## Training the KMeans Model

```python
kmeans = KMeans(n_clusters=5)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

## Visualizing the Entire Data

```python
plt.scatter(df[['Spending Score (1-100)']],df[['Annual Income (k$)']])
plt.xlabel("Spending Score")
plt.ylabel("Annual Income")
plt.title("Mall Customers")
plt.show()
```

## Plotting Individual Clusters


```python
sns.set_style(style='darkgrid')
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1], c="#7C83FD", label="Cluster-1")
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1], c="#FFC947", label="Cluster-2")
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1], c="#28FFBF", label="Cluster-3")
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1], c="#FC5404", label="Cluster-4")
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1], c="#B980F0", label="Cluster-5")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], c="#480032", label="Centroids", s=170)
plt.legend(loc="upper right")
plt.ylabel("Spending Score")
plt.xlabel("Annual Income")
plt.title("Cluster Visualization")
plt.show()
```
